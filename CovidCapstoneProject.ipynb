TODO A - Clean the scraped data

```python
ncdc = ncdc.rename(columns = {'States Affected': 'states',
    'No. of Cases (Lab Confirmed)': 'Number of cases confirmed in Lab',
                 'No. of Cases (on admission)':'Number of cases on admission',
                 'No. Discharged' : 'Number of Patients discharged',
                 'No. of Deaths': 'Number of Deaths'})
ncdc.head()
```

```python
#Dropped the column with null entries
recovered_cases.drop(columns='Province/State',inplace=True)
confirmed_cases.drop(columns='Province/State',inplace=True)
deaths.drop(columns='Province/State',inplace=True)
```

```python
confirmed_cases
```

TODO B - Get a Pandas DataFrame for Daily Confirmed Cases in Nigeria. Columns are Date and Cases

```python
Nigeria = confirmed_cases[confirmed_cases['Country/Region'] == 'Nigeria']
columns = Nigeria.iloc[:, 4:]

nigeria_confirmed = Nigeria.melt(value_vars=columns, var_name="Date", value_name="Cases")
nigeria_confirmed["Date"] = nigeria_confirmed["Date"].apply(pd.to_datetime, errors='coerce')

nigeria_confirmed.head()
```

TODO C - Get a Pandas DataFrame for Daily Recovered Cases in Nigeria. Columns are Date and Cases

```python
Nigeria = recovered_cases[recovered_cases['Country/Region'] == 'Nigeria']
columns = Nigeria.iloc[:, 4:]

nigeria_recovered = Nigeria.melt(value_vars=columns, var_name="Date", value_name="Cases")
nigeria_recovered["Date"] = nigeria_recovered["Date"].apply(pd.to_datetime, errors='coerce')
nigeria_recovered.head()
```

TODO D - Get a Pandas DataFrame for Daily Death Cases in Nigeria. Columns are Date and Cases

```python
Nigeria = deaths[deaths['Country/Region'] == 'Nigeria']
columns = Nigeria.iloc[:, 4:]

nigeria_death = Nigeria.melt(value_vars=columns, var_name="Date", value_name="Cases")
nigeria_death["Date"] = nigeria_death["Date"].apply(pd.to_datetime, errors='coerce')
nigeria_death.head()
```

This DataFrame will contain top 10 states in terms of confirmed cases i.e sort by confirmed cases. ** Hint: Check out Pandas [nlargest](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html) function.
```python
sns.regplot(y = 'Number of cases confirmed in Lab', x = 'Overall CCVI Index', data = NCDC_external)
```

Insights:
There's a negative relationship between number of cases confirmed in Lab and overall CCVI Index that is the higher the number of confirmed cases the lower the overall community vulnerability index in the states except in FCT
